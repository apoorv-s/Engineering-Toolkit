## Laplace Eqn

```python
from DifferentialEquations.Laplace import solve_laplace
# For solving laplace eqn using Finite difference on a rectangular grid
```

### Example 1

```python
def boundary_condition(x, y):
    return (x**2)+(y**2)

x_grid=np.arange(-1, 1+0.1, 0.1)
y_grid=np.arange(-1, 1+0.1, 0.1)
```

```python
# Solve the Laplace equation
u_numerical = solve_laplace(x_grid, y_grid, boundary_condition)
```

```python
plt.imshow(u_numerical.T, origin="lower", extent=[-1, 1, -1, 1])
plt.colorbar()
plt.xlabel("x")
plt.ylabel("y")
plt.title("u(x, y)")
```
